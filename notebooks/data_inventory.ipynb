Metadata-only inventory and consistency report for the shared data store. Set `S2_DATA_STORE` to the local store root; defaults to `~/data-store`.

```python
import os
from pathlib import Path

from s2_msi_raw_generator import inventory

STORE = Path(os.environ.get("S2_DATA_STORE") or "~/data-store").expanduser()
payload = inventory.write_outputs(STORE)
records = payload["records"]
findings = payload["findings"]
print(f"store: {STORE}")
print(f"items: {len(records)}  findings: {len(findings)}")
print(f"wrote: {STORE / 'INVENTORY.md'}")
```

```python
print("kind                           bytes      group                                      path")
print("-" * 110)
for rec in records:
    print(f"{rec['kind']:<30} {rec['bytes']:>10}  {rec.get('group', '')[:40]:<40}  {rec['path']}")
```

```python
print("severity  id                              evidence")
print("-" * 110)
for finding in findings:
    print(f"{finding['severity']:<9} {finding['id']:<31} {finding['evidence']}")
```
