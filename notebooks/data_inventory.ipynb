# Data-store inventory

Metadata-only inventory and consistency report for the shared data store. Set `S2_DATA_STORE` to the local store root; defaults to `~/data-store`.

```python
# One-time kernel setup — safe to re-run. After installing, RESTART the kernel.
import sys

_need_pkg = _need_deps = False
try:
    import s2_msi_raw_generator  # noqa: F401
except Exception:
    _need_pkg = True
try:
    import zarr  # noqa: F401
except Exception:
    _need_deps = True
if _need_pkg:
    !{sys.executable} -m pip install --quiet -e .. --no-deps
if _need_deps:
    !{sys.executable} -m pip install --quiet zarr
    !{sys.executable} -m pip install --quiet --force-reinstall --no-deps zarr
if _need_pkg or _need_deps:
    print("installed/repaired — RESTART the kernel and run all cells again")
else:
    print("environment OK")
```

```python
import os
from pathlib import Path

from s2_msi_raw_generator import inventory

STORE = Path(os.environ.get("S2_DATA_STORE") or "~/data-store").expanduser()
payload = inventory.write_outputs(STORE)
records = payload["records"]
findings = payload["findings"]
print(f"store: {STORE}")
print(f"items: {len(records)}  findings: {len(findings)}")
print(f"wrote: {STORE / 'INVENTORY.md'}")
```

```python
print("kind                           bytes      group                                      path")
print("-" * 110)
for rec in records:
    print(f"{rec['kind']:<30} {rec['bytes']:>10}  {rec.get('group', '')[:40]:<40}  {rec['path']}")
```

```python
print("severity  id                              evidence")
print("-" * 110)
for finding in findings:
    print(f"{finding['severity']:<9} {finding['id']:<31} {finding['evidence']}")
```
